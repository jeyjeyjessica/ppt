```python
import networkx as nx
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(2,3)
nx.draw_networkx(G, with_labels=True)
```

```python
G = nx.Graph()
G = nx.gnp_random_graph(12,0.5)
nx.draw_networkx(G,with_labels=True)
```

```python
print(len(G.edges()))
```

```python
print(len(G.nodes()))
```

```python
nx.degree(G)
```

```python
dc = nx.degree_centrality(G)
print(dc)
```

```python
for w in sorted(dc, key=dc.get, reverse=True):
  print(w, dc[w])
```

```python
bc = nx.betweenness_centrality(G)
for w in sorted(bc, key=bc.get, reverse=True):
  print(w, bc[w])
```

```python
ec = nx.eigenvector_centrality(G)
for w in sorted(ec, key=ec.get, reverse=True):
  print(w, ec[w])
```

```python
cc = nx.closeness_centrality(G)
for w in sorted(cc, key=cc.get, reverse=True):
  print(w, cc[w])
```

```python
nx.clustering(G)
```

```python
bridges = list(nx.bridges(G))
print(len(bridges))
```

```python
tri = list(nx.triangles(G))
print(len(tri))
```

```python
local = list(nx.local_bridges(G))
print(len(local))
```
