```python
!pip install cartopy
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# Load your dataset
df = pd.read_csv('/content/MOCK_DATA (8).csv')

# Basic statistics
#print(df.describe())

# Plotting tweets on a map
plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()
ax.scatter(df['longitude'], df['latitude'], color='red', alpha=0.7, transform=ccrs.PlateCarree())
ax.set_title('Tweets Distribution')
plt.show()

# Plotting tweet density by city
city_counts = df['city'].value_counts()
plt.figure(figsize=(10, 6))
city_counts.plot(kind='bar')
plt.title('Tweet Density by City')
plt.xlabel('City')
plt.ylabel('Number of Tweets')
plt.show()
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv('your_dataset.csv')

# Split the combined column into separate longitude and latitude columns
df[['longitude', 'latitude']] = df['combined_column'].str.split(',', expand=True).astype(float)

# Drop the combined column if no longer needed
df.drop(columns=['combined_column'], inplace=True)

# Now you can perform your analysis
# For example, you can plot tweets on a map
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()
ax.scatter(df['longitude'], df['latitude'], color='red', alpha=0.7, transform=ccrs.PlateCarree())
ax.set_title('Tweets Distribution')
plt.show()

# Or you can analyze tweet density by city
city_counts = df['city'].value_counts()
plt.figure(figsize=(10, 6))
city_counts.plot(kind='bar')
plt.title('Tweet Density by City')
plt.xlabel('City')
plt.ylabel('Number of Tweets')
plt.show()
```
